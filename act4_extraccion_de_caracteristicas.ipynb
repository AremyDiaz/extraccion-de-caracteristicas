```python
#Importar librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from funpymodeling.exploratory import freq_tbl
```

```python
pd.options.display.max_columns=None
```

```python
#Leer excel
df_microretailer=pd.read_excel('microretailer_mit_lift_lab.xlsx')
df_microretailer.head()
```

```python
df_microretailer.info()
```

```python
#Selección de variables categóricas
#Numéricas y cualitativas 2020
cualitativas_20=df_microretailer[['_updated_by','232_type_of_store','108_does_the_micro_retailer_has_a_barred_window_','99_does_the_micro_retailer_exhibits_products_outside_','102_does_the_micro_retailer_sells_fresh_products_','28_prefered_payment_method','30_payment_methods_large_purchases','214_customer_relationship_tools','33_credit_to_customers','193_sales_planning_tools','64_comission_services','313_home_deliveries','185_place_orders_suppliers','53_special_conditions_suppliers','192_procurement_planning_tools','277_payment_method_suppliers','56_procurement_changes_pandemic','157_frequency_organize_shelves','158_charact_organize_products_on_shelves','234_willingness_participate_workshop']]
cualitativas_20
```

```python
#Limpiar nulos del df completo
cualitativas_20=cualitativas_20.fillna('Unknown')
nulos_cualitativas_20=cualitativas_20.isnull().sum().sum()
nulos_cualitativas_20#Comprobación
```

```python
#Como son categóricas, no necesita limpieza de outliers
```

Columna 1: _updated_by

```python
#Análisis univariado de una sola columna
table1=freq_tbl(cualitativas_20['_updated_by'])
table1
```

```python
#Filtrar cumulative_perc mayor al 50%
filtro1=table1[table1['cumulative_perc']<0.55]
filtro1
```

Columna 2: 232_type_of_store

```python
#Análisis univariado de una sola columna
table2=freq_tbl(cualitativas_20['232_type_of_store'])
table2#64 categorías
```

```python
#Filtrar frecuencia al menos de 5
filtro2=table2[table2['frequency']>=5]
filtro2
```

```python
#Poner las categorías como índices
filtro2_index=filtro2.set_index('232_type_of_store')
filtro2_index
```

```python
#Barplot con el df filtrado
filtro2_index['frequency'].plot(kind="bar",width=0.7,figsize=(10,4))
plt.title('type_of_store frequency')
plt.xlabel('type_of_store')
plt.ylabel('frequency')
```
